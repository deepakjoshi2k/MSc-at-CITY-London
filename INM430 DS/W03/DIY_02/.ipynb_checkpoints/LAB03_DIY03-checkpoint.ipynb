```python
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

sample_sizes = [5, 10, 100]
std_devs = [1, 2, 5]

# Storage for results
results = []

for s in std_devs:
    for n in sample_sizes:
        sample = np.random.normal(loc=0, scale=s, size=n)
        
        mean_val = np.mean(sample)
        std_val = np.std(sample, ddof=1)
        skew_val = skew(sample)
        kurt_val = kurtosis(sample)
        
        results.append({
            'n': n,
            'std_dev (true)': s,
            'sample_mean': mean_val,
            'sample_std': std_val,
            'sample_skewness': skew_val,
            'sample_kurtosis': kurt_val
        })

# Convert to DataFrame
results_df = pd.DataFrame(results)
results_df
```

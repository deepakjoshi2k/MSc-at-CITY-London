```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('censusCrimeClean.csv')
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
medInc=df[['medIncome']]
VioCrimPP=df.ViolentCrimesPerPop
```

```python
sns.scatterplot(data=df,x=df.medIncome,y=VioCrimPP)
```

```python
from sklearn.linear_model import LinearRegression

Lr=LinearRegression().fit(medInc,VioCrimPP)
```

```python
Lr.coef_[0]
```

```python
Lr.intercept_
```

```python
Lr.score(medInc,VioCrimPP)
```

```python
modelledDependent = np.polyval([Lr.coef_[0], Lr.intercept_], df.medIncome)
```

```python
modelledDependent.min()
```

```python
newdf=pd.DataFrame(columns=['medIncome','dep'])
```

```python
newdf.medIncome=df.medIncome
newdf.dep=modelledDependent
newdf
```

```python
df.medIncome.shape
```

```python
sns.scatterplot(data=df,x=df.medIncome,y='ViolentCrimesPerPop')
sns.scatterplot(data=newdf,x=df.medIncome,y='dep')
```

##### Trying 2d order Poly

```python
df
```

```python
pCoeff = np.polyfit(df.medIncome, df.ViolentCrimesPerPop, 2)
evaluatedCurve = np.polyval(pCoeff, df.medIncome)

n_df=pd.DataFrame(columns=['medIncome','dep'])
n_df.medIncome=df.medIncome
n_df.dep=evaluatedCurve
```

```python
sns.scatterplot(data=df,x=df.medIncome,y='ViolentCrimesPerPop')
sns.scatterplot(data=n_df,x=df.medIncome,y='dep')
```

#### Comparing

```python
sns.scatterplot(data=df,x=df.medIncome,y='ViolentCrimesPerPop')
sns.scatterplot(data=newdf,x=df.medIncome,y='dep',color='red')
sns.scatterplot(data=n_df,x=df.medIncome,y='dep',color='yellow')
```

# T1 Merge the two files based on the column they share.

```python
df_mergerd=df1.merge(df2,on='TicketType',how='inner')
df_mergerd.head()
```

# T2. Display the name of the oldest passengers (hint: make use of variables to save some intermediate values).

```python
old_p=df_mergerd[df_mergerd.Age==np.max(df_mergerd.Age)]
old_p
```

# T3. Plot the data on a scatter plot that shows the Age vs. Ticket Prices

```python
sns.scatterplot(data=df_mergerd,x='Age',y='Fare')
```

# T4. Plot only the data that shows female passengers aged 40 to 50 and who paid more than or equal to 40.

```python
condition=df_mergerd[((df_mergerd.Sex=='female') & (df_mergerd.Fare>=40)) & ((df_mergerd.Age>=40) & (df_mergerd.Age<=50))]
sns.scatterplot(condition,x='Age',y='Fare')
```

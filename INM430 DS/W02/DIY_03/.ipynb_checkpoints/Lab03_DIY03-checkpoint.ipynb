```python
sns.histplot(data=df,x=df.e_tbhiv_prct_lo)
plt.title("Original Distribution")
plt.savefig("Original Distribution",facecolor='white')
```

## Impact of median

```python
df1=df.copy(deep=True)
m=df1.e_tbhiv_prct_lo.median()
df1.e_tbhiv_prct_lo.fillna(m,inplace=True)
```

```python
sns.histplot(data=df1,x=df1.e_tbhiv_prct_lo)
plt.title("Impact of Median (Imputation)")
plt.savefig("Median Distribution",facecolor='white')
```

## Impact of mean

```python
df2=df.copy(deep=True)
m1=df.e_tbhiv_prct_lo.mean()
df2.e_tbhiv_prct_lo.fillna(m1,inplace=True)
sns.histplot(data=df2,x=df2.e_tbhiv_prct_lo)
plt.title("Impact of Mean (Imputation)")
plt.savefig("Mean Distribution",facecolor='white')
```

## Impact of Interpolation

```python
df3=df.copy(deep=True)
df3.e_tbhiv_prct_lo.fillna(method="ffill",inplace=True) #fill same value forward untill vaild value comes
sns.histplot(data=df3,x=df3.e_tbhiv_prct_lo)
plt.title("Impact of forward fill (Imputation)")
plt.savefig("Forward fill Distribution",facecolor='white')
```

```python
df4=df.copy(deep=True)
df4.e_tbhiv_prct_lo.fillna(method="bfill",inplace=True) #fill same value backward untill vaild value comes
sns.histplot(data=df4,x=df4.e_tbhiv_prct_lo)
plt.title("Impact of backward fill (Imputation)")
plt.savefig("Backward fill Distribution",facecolor='white')
```

```python
df5=df.copy(deep=True)
df5.e_tbhiv_prct_lo.interpolate(meathod="linear",inplace=True)
sns.histplot(data=df5,x=df5.e_tbhiv_prct_lo)
plt.title("Impact of Linear Imputation")
plt.savefig("Linear Imputation",facecolor='white')
```

```python
plt.tight_layout()
plt.figure(figsize=(10,12))


plt.subplot(3,2,1)
plt.tight_layout()
sns.histplot(data=df,x=df.e_tbhiv_prct_lo,bins="doane")
plt.title("Original Distribution")
plt.ylim(0,3500)

plt.subplot(3,2,2)
plt.tight_layout()
sns.histplot(data=df5,x=df5.e_tbhiv_prct_lo,bins="doane")
plt.title("Imact of Linear Interpolation")
plt.ylim(0,3500)

plt.subplot(3,2,3)
plt.tight_layout()
sns.histplot(data=df1,x=df1.e_tbhiv_prct_lo,bins="doane")
plt.title("Impact of Median (Imputation)")
plt.ylim(0,3500)

plt.subplot(3,2,4)
plt.tight_layout()
sns.histplot(data=df2,x=df2.e_tbhiv_prct_lo,bins="doane")
plt.title("Impact of Mean (Imputation)")
plt.ylim(0,3500)

plt.subplot(3,2,5)
plt.tight_layout()
sns.histplot(data=df3,x=df3.e_tbhiv_prct_lo,bins="doane")
plt.title("Impact of forward fill (Imputation)")
plt.ylim(0,3500)

plt.subplot(3,2,6)
plt.tight_layout()
sns.histplot(data=df4,x=df4.e_tbhiv_prct_lo,bins="doane")
plt.title("Impact of backward fill (Imputation)")
plt.ylim(0,3500)

plt.suptitle("COMPARISON OF TECHNIQUES",fontsize=16)
plt.tight_layout()
plt.savefig("Basic Imputation",facecolor='white')

plt.show()
```

## Using "Linear Interpolation and backward fill" to fill the missing values.

```python
df_try=df.copy(deep=True)
df_try.interpolate(method='linear',inplace=True)
df_try.fillna(method='bfill', inplace=True)
```

```python
sns.histplot(data=df_try,x=df_try.e_tbhiv_prct_lo,bins="doane")
plt.title("Impact of backward fill (Imputation)")
plt.ylim(0,3500)
```
